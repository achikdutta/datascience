-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Write your code for extracting the first two genres of each movie here

## --- Split the genres column by | also create genres_XXX column fro each split --- | START
movies_genres=movies_r_null.join(movies_r_null['genres'].str.split('|',expand=True).add_prefix('genres_'))
## --- Split the genres column by | also create genres_XXX column fro each split --- | END

## --- Drop the other columns other than genres_0 and genres_1 --- | START
movies_genres=movies_genres.drop(['genres_2','genres_3','genres_4','genres_5','genres_6','genres_7'],axis=1)
## --- Drop the other columns other than genres_0 and genres_1 --- | END

## --- Rename the columns genres_0 and genres_1 to genres_1 and genres_2 respectively --- | START
movies_genres=movies_genres.rename(columns={'genres_0':'genres_1','genres_1':'genres_2'})
## --- Rename the columns genres_0 and genres_1 to genres_1 and genres_2 respectively --- | END

## --- Replace None values with NaN values --- | START
movies_genres.genres_2.fillna(value=pd.np.nan, inplace=True)
## --- Replace None values with NaN values --- | END

## --- Replace Null values in genres_2 column with genres_1 column value --- | START
movies_genres.loc[pd.isnull(movies_genres['genres_2']),['genres_2']]=movies_genres['genres_1']
## --- Replace Null values in genres_2 column with genres_1 column value --- | END
```

```python
## --- Group by genres_1 and genres_2 --- | START
movies_by_segment = movies_genres.groupby(['genres_1','genres_2'])
## --- Group by genres_1 and genres_2 --- | END
```

```python
## --- Use the groupby object and take mean of the gross, convert the Series into dataframe, sort by gross value, take the first 5 rows --- | START --- | START
PopGenre = ((movies_by_segment['gross'].mean()).reset_index().sort_values(by="gross",ascending=False)).head(5)
## --- Use the groupby object and take mean of the gross, convert the Series into dataframe, sort by gross value, take the first 5 rows --- | END

## Family + Sci-Fi is the highest with max gross value
```
